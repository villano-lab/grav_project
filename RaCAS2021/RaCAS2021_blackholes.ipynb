```python
from sympy.utilities.lambdify import lambdify
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive, fixed, FloatSlider, HBox, Layout, Button, Label, Output, VBox
from IPython.display import display

w=3970 #width of the square image
h=w
#center of galaxy:
c_x=w/2
c_y=h/2.2

#inner/outer radii, judging by eye
r1=.16*w
r2=.28*w

def f(arraysize,Mbh):
    #Mbh=3 #Msun, arbitarily chosen b/t minmass and maxmass [Msun]
    minkpc=0
    maxkpc=25
    kpctopixels=60
    scale=.2e9 #somewhat arbitrary
    rc=2.4 #kpc for ngc5005...
    rho00=.31e9 #Msun/kpc^2 for ngc5533
    r = symbols('r')
    X=r/rc
    Z = rho00/(X*(1+X)**2) #NFW (dark halo) density profile. I'm assuming Rs is cutoff radius. is this correct?
    RHO=4*np.pi*r**2*Z #
    RRR=integrate(RHO)
    func = lambdify(r, RRR,'numpy') # returns a numpy-ready function

    rr=np.linspace(minkpc,maxkpc,int(arraysize)) #kp
    M = func(rr) #c prepresents the total mass [Msun] at each point in the array rr [kpc]

    
    Nbh=M/Mbh #number of hypothetical tiny black holes that would have to exist b/t minkpc and maxkpc
    
    N=M/scale #the actual number of lil blackholes each projected dot represents
    N=np.rint(N) #convert array to integers
    maxN=max(N)
    scaled=N/maxN
    length=len(scaled)
    brackets=5
    brackarraysize=int(length/brackets)
    bracket1=N[0:brackarraysize]
    N1=np.mean(bracket1)
    bnum=np.linspace(1,brackets,brackets)
    r2=np.rint(bnum*brackarraysize)
    r1=r2+1-brackarraysize
    r1=r1*kpctopixels
    r2=r2*kpctopixels
    r1a,r1b,r1c,r1d,r1e=r1
    r2a,r2b,r2c,r2d,r2e=r2
    r1aa=np.rint(r1-1)
    r2aa=np.rint(r2-1)
    r2ax,r2bx,r2cx,r2dx,r2ex=r2aa
    r1ax,r1bx,r1cx,r1dx,r1ex=r1aa
    r1ax=int(r1ax)
    r1bx=int(r1bx)
    r1cx=int(r1cx)
    r1dx=int(r1dx)
    r1ex=int(r1ex)
    r2ax=int(r2ax)
    r2bx=int(r2bx)
    r2cx=int(r2cx)
    r2dx=int(r2dx)
    r2ex=int(r2ex)
    bracketa=N[r1ax:r2ax]
    bracketb=N[r1bx:r2bx]
    bracketc=N[r1cx:r2cx]
    brackete=N[r1ex:r2ex]
    bracketd=N[r1dx:r2dx]
    Na=np.mean(bracketa)
    Nb=np.mean(bracketb)
    Nc=np.mean(bracketc)
    Nd=np.mean(bracketd)
    Ne=np.mean(brackete)
    ra = np.random.uniform(r1a,r2a,int(arraysize))
    rb = np.random.uniform(r1b,r2b,int(arraysize))
    rc = np.random.uniform(r1c,r2c,int(arraysize))
    rd = np.random.uniform(r1d,r2d,int(arraysize))
    re = np.random.uniform(r1e,r2e,int(arraysize))
    
    angle=np.random.uniform(0,2*np.pi,int(arraysize)) #angle 0 to 360 degrees for fulle circle (donut)
    ya=c_y+ra*np.sin(angle) #y coordinates
    xa=c_x+ra*np.cos(angle) #x coordinates
    yb=c_y+rb*np.sin(angle) #y coordinates
    xb=c_x+rb*np.cos(angle) #x coordinates
    yc=c_y+rc*np.sin(angle) #y coordinates
    xc=c_x+rc*np.cos(angle) #x coordinates
    yd=c_y+rd*np.sin(angle) #y coordinates
    xd=c_x+rd*np.cos(angle) #x coordinates
    ye=c_y+re*np.sin(angle) #y coordinates
    xe=c_x+re*np.cos(angle) #x coordinates

    img = plt.imread("A_spiral_snowflake.jpg") #import special snowflake ngc 6814, which has visual diameter about 27.6kpc
    plt.figure(figsize=(20,24))
    size=5*Mbh #display size of each marker
    plt.scatter(xa,ya,color="k",marker='.',s=size)
    plt.scatter(xb,yb,color="k",marker='.',s=size)
    plt.scatter(xc,yc,color="k",marker='.',s=size)
    plt.scatter(xd,yd,color="k",marker='.',s=size)
    plt.scatter(xe,ye,color="k",marker='.',s=size)
    #plt.scatter(0,100, color="blue")
    #plt.scatter(c_x,c_y,color="red")
    plt.imshow(img)
    plt.axis('off')
   
style = {'description_width': 'initial'}
layout = {'width':'600px'}

################################
######## Define Sliders ########
################################


Max=100
best_M=.5*Max
stepN=.1*Max
Min=1

#for blackhole mass slider:
minmass=.1 #solar masses, arbitrary
maxmass=3.7 #solar masses, just smaller then the smallest black hole ever discovered
start=.5*maxmass

#mass of each black hole
Mbh = FloatSlider(min=minmass, max=maxmass, step=minmass, 
                value=start,
                description='Mass of each lil black holes [Msun]', 
                readout= True,
                readout_format='.1f', 
                orientation='horizontal', 
                style=style, layout=layout)

#number of projected black dots slider
arraysize = FloatSlider(min=Min, max=Max, step=stepN, 
                value=best_M, 
                description='Number of lil black holes multiplied by scale %.0e'%scale, 
                readout= True,
                readout_format='.2d', 
                orientation='horizontal', 
                style=style, layout=layout)



def interactive_plot(f):
    interact = interactive(f, arraysize=arraysize,Mbh=Mbh,continuous_update=False)
    return interact
VBox([interactive_plot(f)])
```

```python
#GETTING AN ESTIMATE OF HOW MANY BLACK HOLES EACH PROJECTED DOT REPRESENTS

from scipy.integrate import quad
import numpy as np
rc=2.4 #kpc
rho00=.31e9 #Msun/kpc^2


N=100 #number of blackholes visually projected onto image
#inner and outer radius of projected lil black holes, 
r1=10 #kpc, reasonable values visually given NGC6814's diameter
r2=15


def rho(r):
    X=r/rc
    Z = rho00/(X*(1+X)**2) #NFW (dark halo) density profile. I'm assuming Rs is cutoff radius. is this correct?
    return 4*np.pi*r**2*Z #and this is the integral to find the mass according to wikipedia
#https://en.wikipedia.org/wiki/Navarro%E2%80%93Frenk%E2%80%93White_profile
Mg = quad(rho, r1, r2)[0] #total mass, in Msun, b/t r1 and r2

Mbh=3 #Msun, mass smaller than smallest black hole yet discovered
Nbh=Mg/Mbh #number of hypothetical tiny black holes that would have to exist b/t r1 and r2

scale=round(Nbh/N,0)#the actual number of lil blackholes each projected dot represents
print('each black dot represents %.0f lil black holes' %scale)
```
